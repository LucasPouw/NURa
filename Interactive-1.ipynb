```python
import numpy as np

class Interpolator():

    def __init__(self,
                 points):
        
        '''Input points must be (2, N) array'''

        self.points = points
        self.points_x = points[0, :]  # TODO: Assert that the points are strictly increasing by checking and making decreasing into increasing by * -1
        self.points_y = points[1, :]
        

    def bisection(self, x, order):
        '''Find the closest point on either side of x. This implementation only works for an INCREASING set of data points.'''

        left, right = 0, len(self.points_x) - 1  # Starting indeces are the first and last point

        while right - left > 1:

            mid = (left + right) // 2
            
            if x <= self.points_x[mid]:  # We choose the left side of the interval to include the boundary
                right = mid
            else:
                left = mid

        n_points_to_use = order // 2
        lowest_idx = left - n_points_to_use + 1  # We default to using less points on the left of x if the order is odd

        if lowest_idx < 0:
            print(f'WARNING: The requested order of {order} requires {n_points_to_use} points on the left of the queried \
value. Found {n_points_to_use + lowest_idx}. Getting the remainder from the right side of the query.')
            lowest_idx = 0

        return lowest_idx
    

    def linear(self, array):

        assert len(self.points) >= 2, f'Linear interpolation requires at least 2 points. Found {len(self.points)}'

        interpolated = np.zeros(len(array))
        for i, x in enumerate(array):
            lowest_idx = self.bisection(x, order=2)
            highest_idx = lowest_idx + 1

            x_low, y_low = self.points[:, lowest_idx]
            x_high, y_high = self.points[:, highest_idx]

            interpolated[i] = (y_high - y_low) * float(x_high - x_low)**(-1) * (x - x_low) + y_low

        return interpolated
    
    
    def polynomial(self, array, order):

        assert len(self.points) >= order, f'Interpolation of order {order} requires at least {order} points. Found {len(self.points)}'

        return
    
if __name__ == '__main__':

    import matplotlib.pyplot as plt
    from matplotlib.image import imread

    image = imread("M42_128.jpg")
    first_row_pixvals = image[0,:]
    first_row_pixidx = np.arange(image.shape[0])
    
    data = np.array([first_row_pixidx, first_row_pixvals])

    Interp = Interpolator(data)

    xinterp = np.linspace(0, image.shape[0], 201)
    yinterp = Interp.linear(xinterp)

    plt.figure(figsize=(8,6))
    plt.plot(data[0,:], data[1,:], color='red')
    plt.plot(xinterp, yinterp, color='blue')
    plt.scatter(xinterp, yinterp, marker='o', color='black', zorder=5)
    plt.show()
```
